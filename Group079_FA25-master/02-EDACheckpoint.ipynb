# Research Question

How have San Diego’s drinking and tap water quality levels (measured by the presence of chemical parameters: fluoride, THM, ton, bromide, etc.) changed over time, and are these changes associated with changes in reported public health outcomes (measured by occurrences of specific cancers such as brain cancer, lung cancer, prostate cancer, etc.) in those areas?

## Background and Prior Work

Clean water is one of the most complicated, yet crucial public health challenges. While treatment and better infrastructure have advanced, chemical pollutants still manage to slip into systems that people use every day. Substances such as trihalomethanes (THMs), bromides, and traces of heavy metals can make their way into tap water through everything from old pipes and industrial discharge to agricultural runoff after heavy rain.<a href="#ref1">¹</a> Repeated exposure, especially at low but consistent levels over time, has been linked with chronic diseases, including several types of cancer such as leukemia and colorectal cancer. The U.S. Environmental Protection Agency (EPA, 2024) has made it clear that ensuring safe drinking water is not just a matter of meeting the numbers on a chart but also about reducing toxic exposure that can silently build up over years.<a href="#ref2">²</a>  

In the context of San Diego, although the water supply of this city generally meets federal standards, that does not mean that it's perfectly safe in the long run. Those standards reflect only the latest findings about long-term exposure risks, some researchers point out. For example, chemicals such as THMs and bromides can turn problematic during long-term exposure. And here's where things could become problematic: depending on the source, treatment method, or even plumbing conditions in various neighborhoods, the concentration of these contaminants can vary considerably across town. A question that naturally follows: over time, have the fluctuations of those chemicals tracked with fluctuations in local cancer statistics? That may be a difficult connection to prove, yet investigating it offers insight at how environmental conditions and health outcomes interconnect in everyday life.  

There is some preliminary work that is worth noting. Ren, Martinez, and Cohen (2024) conducted a national-level analysis and found that areas with more Safe Drinking Water Act violations tended to report poorer physical and mental health—along with higher healthcare costs.<a href="#ref3">³</a> Liu, Zhang, and Wang (2024) looked at water pollution in China and found people with access to cleaner water supplies were far more likely to report their health as “very good.”<a href="#ref4">⁴</a> These are good studies that reinforce what seems intuitive but needs to be reinforced with data: water quality affects health. Yet, studies that link the specific chemistry of a city's water system—things like fluoride levels, THMs, and bromide—to concrete health outcomes, like cancer rates, still need to be done. By sifting through San Diego's water data over time and comparing it to public health statistics, this research hopes to find patterns that may go otherwise unseen. This research is important because it helps uncover how changes in San Diego’s water quality may be quietly affecting community health.

# Hypothesis


**Hypothesis**: We predict that there is a positive correlation between the presence of chemical contaminants in San Diego’s drinking water supply and the incidence rates of cancer reported in the same areas over time.

**Null Hypothesis**: There is no statistically significant correlation between the presence of chemical contaminants in San Diego’s drinking water supply and the incidence rates of cancer reported in the same areas over time.

**Rationale**: Research has shown that excessive exposure to chemical contaminants found in drinking water - such as from manufacturing processes, agricultural runoff, and hard metals - has led to an increased risk in developing chronic diseases, including cancer. In San Diego’s tap water, chemicals like THMs and bromides have been found and are known to be associated with having carcinogenic effects with long-term exposures. Therefore, monitoring how levels of chemical contaminants have changed over time in water in San Diego may reveal the positive correlation between the presence of chemical contaminants and trends in local cancer occurrence.

#### Section 1: Variable Distributions Across All Datasets

This section aims to explore the distributions of key variables in all three datasets used in our project:

- Water Treatment Plant Effluent Dataset (8,560 rows, 8 variables)

- Water Distribution System Dataset (29,377 rows, 8 variables)

- Community Morbidity & Mortality (Cancer) Dataset (5,828 rows, 6 variables)

- The goal is to understand the shape of our data (normal vs skewed), identify heavily sampled vs rarely sampled variables, and evaluate whether the data is clean enough for further analysis.

Understanding distributions is important because right-skewed variables, rare categories, and uneven sampling influence our ability to merge datasets and draw inferences.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load cleaned datasets from Checkpoint 1
effluent = pd.read_csv("./data/02-processed/clean_effluent_dataset.csv")
distribution = pd.read_csv("./data/02-processed/clean_distribution_dataset.csv")
cancer = pd.read_csv("./data/02-processed/clean_cancer_dataset_2016_2022.csv")

print(effluent.head(), distribution.head(), cancer.head())

print("effluent shape:" , effluent.shape)
print("distribution shape:" , distribution.shape)
print("cancer shape:" , cancer.shape)
```

Pandas's read_csv makes all columns string by default, so we need to convert the date_sample column to datetime.

```python
effluent['date_sample'] = pd.to_datetime(effluent['date_sample'])
distribution['date_sample'] = pd.to_datetime(distribution['date_sample'])
cancer['Year'] = pd.to_datetime(cancer['Year'])
```

##### Water Treatment Plant Effluent Dataset — Distributions

- date_sample: Evenly distributed from 2016–2025 with one gap in December 2020.

- analyte: Fluoride, pH, Total THMs, Chlorate, Chlorite, all have substantial samples. Fluoride has slightly more, and Chlorite and Chlorate slightly less, with a difference of ~7x between most and least sampled.

- analyte_value: FFluoride, pH, and Chlorate are fairly normally distributed. Chlorite has a right tail. Total THMs has a long right tail with some extreme outliers, but still shows right skew if they are removed.

- sample_source: 35 sources, with a good distribution between them. There are a few more samples from OT Filters than the rest. These map 1-to-1 with source_description, so we don't need to analyze both.

```python
plt.figure(figsize=(15, 6))

effluent_monthly_plot = effluent['date_sample'].dt.to_period('M').value_counts().sort_index()

plt.bar(range(len(effluent_monthly_plot)), effluent_monthly_plot.values)
plt.title('Effluent Samples by Year-Month', fontsize=14, fontweight='bold')
plt.ylabel('Number of Samples', fontsize=12)
plt.xlabel('Year-Month', fontsize=12)

tick_positions = range(0, len(effluent_monthly_plot), 6)
tick_labels = [str(effluent_monthly_plot.index[i]) for i in tick_positions]
plt.xticks(tick_positions, tick_labels, rotation=45, ha='right')

plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(data=effluent, x='analyte', order=effluent['analyte'].value_counts().index)
plt.title("Frequency of Analytes (Effluent Dataset)")
plt.show()
```

```python
analytes = effluent['analyte'].unique()
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 8))
axes = axes.flatten()

for i, analyte in enumerate(analytes):
    analyte_data = effluent[effluent['analyte'] == analyte]
    
    # Plot on the specific subplot
    sns.histplot(data=analyte_data, x='analyte_value', bins=50, kde=True, ax=axes[i])
    axes[i].set_title(f'{analyte} Distribution')
    axes[i].set_xlabel('Analyte Value')
    axes[i].set_ylabel('Frequency')


# Look at THMs without outliers
thm_no_outlier = effluent[(effluent['analyte'] == 'Total THMs') & (effluent['analyte_value'] < 100) & (effluent['analyte_value'] > 0)]
sns.histplot(data=thm_no_outlier, x='analyte_value', bins=50, kde=True, ax=axes[5])
axes[5].set_title('Total THMs Distribution (Without Suspicious entries and 0s)')
axes[5].set_xlabel('Analyte Value')
axes[5].set_ylabel('Frequency')

plt.tight_layout()
plt.suptitle('Analyte Value per Analyte (Effluent Dataset)', y=1.02, fontsize=20)
plt.show()
```

```python
analytes = effluent['analyte'].unique()

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 8))
axes = axes.flatten()

for i, analyte in enumerate(analytes):
    analyte_data = effluent[effluent['analyte'] == analyte]

    axes[i].scatter(analyte_data['date_sample'], analyte_data['analyte_value'], alpha=0.6)
    axes[i].set_title(f'{analyte} over Time')
    axes[i].set_xlabel('Date Sampled')
    axes[i].set_ylabel('Analyte Value')

plt.tight_layout()
plt.suptitle('Analyte Value over Time per Analyte (Effluent Dataset)', y=1.02, fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(data=effluent, x='sample_source', order=effluent['sample_source'].value_counts().index)
plt.title("Frequency of Sources (Effluent Dataset)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The effluent analyte values show classic right-skew patterns, suggesting that occasional spikes occur. This is either true contamination events, or instrument/recording errors. The analyte frequencies confirm that most advanced analytes are rarely measured, which limits multi-analyte comparisons.

## Ethics

The data we are using does not unfortunately represent every neighborhood in San Diego; it’s biased towards more developed neighborhoods, meaning that some other communities within San Diego may be underrepresented. The main issue we find with the data we proposed is that, despite being generally good at showcasing the state of San Diego’s water quality as a whole city, it is not good at accurately representing the disparities in water quality across all San Diego residents.An issue that comes about from biases when sampling is that the health dataset is only representative from those who chose or were even able to seek medical care, meaning that certain communities or individuals with limited access to healthcare are going to be underrepresented in the data. And, unfortunately, this portion of the population would also most likely be the ones who similarly have limited access to cleaner water and thus may face greater environmental/health risks overall, risking our data’s integrity. Our data may be good at representing the whole population, but it is at risk of overrepresenting the San Diego residents with better access to healthcare and clean water.
To address the issues regarding bias and problematic data collection, we plan to detect and minimize it all the most before the data analysis. Before starting, we will comb through the water quality dataset and closely examine where the water samples were collected. After compiling the locations, we will compare it to demographic data (such as income level and healthcare access) to find gaps within the data samples. And during the actual analysis, we plan to look for patterns that may be caused by the uneven sampling or missing data. After identifying if it makes an impact, we’ll communicate our findings transparently in our report. We will acknowledge which populations are overrepresented or underrepresented, use data visualizations to make it clear what these limitations are and how it affects our findings. This ensures that, regardless of the sample biases, our findings are accurate and respectful towards all communities represented within the data. In addition to bias, we are aware that privacy is a concern for our datasets in our analysis. Privacy is another valid concern in our datasets, because even though the data is public, some of the water sample data is tied to particular sampling locations, whereas some of the health data is tied to specific communities. With these two datasets, it does risk unintentionally exposing sensitive information about smaller, more vulnerable populations. It could unfairly stigmatize lower-income communities, which paints an unfair picture of their community. We plan to avoid this by grouping our results into larger regions so no single community can be singled out. We also plan to clearly communicate that differences in our findings can be attributed to external factors (such as outdated water infrastructure, uneven access to healthcare, etc) rather than the characteristics of residents in those areas. This approach should minimize harmful interpretations of data while maximizing how much information we can meaningfully share.

And, unfortunately, this portion of the population would also most likely be the ones who similarly have limited access to cleaner water and thus may face greater environmental/health risks overall, risking our data’s integrity. Our data may be good at representing the whole population, but it is at risk of overrepresenting the San Diego residents with better access to healthcare and clean water.

### How will you set out to detect these specific biases before, during, and after/when communicating your analysis?
To address the issues regarding bias and problematic data collection, we plan to detect and minimize it all the most before the data analysis. Before starting, we will comb through the water quality dataset and closely examine where the water samples were collected. After compiling the locations, we will compare it to demographic data (such as income level and healthcare access) to find gaps within the data samples. And during the actual analysis, we plan to look for patterns that may be caused by the uneven sampling or missing data. After identifying if it makes an impact, we’ll communicate our findings transparently in our report. We will acknowledge which populations are overrepresented or underrepresented, use data visualizations to make it clear what these limitations are and how it affects our findings. This ensures that, regardless of the sample biases, our findings are accurate and respectful towards all communities represented within the data.

### Are there any other issues related to your topic area, data, and/or analyses that are potentially problematic in terms of data privacy and equitable impact? How will you handle issues you identified?
Privacy is another valid concern in our datasets, because even though the data is public, some of the water sample data is tied to particular sampling locations, whereas some of the health data is tied to specific communities. With these two datasets, it does risk unintentionally exposing sensitive information about smaller, more vulnerable populations. It could unfairly stigmatize lower-income communities, which paints an unfair picture of their community. We plan to avoid this by grouping our results into larger regions so no single community can be singled out. We also plan to clearly communicate that differences in our findings can be attributed to external factors (such as outdated water infrastructure, uneven access to healthcare, etc) rather than the characteristics of residents in those areas. This approach should minimize harmful interpretations of data while maximizing how much information we can meaningfully share.
