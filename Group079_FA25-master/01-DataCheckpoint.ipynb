## Research Question

How have San Diego’s drinking and tap water quality levels (measured by the presence of chemical parameters: fluoride, THM, ton, bromide, etc.) changed over time, and are these changes associated with changes in reported public health outcomes (measured by occurrences of specific cancers such as brain cancer, lung cancer, prostate cancer, etc.) in those areas?

## Background and Prior Work

Clean water is one of the most complicated, yet crucial public health challenges. While treatment and better infrastructure have advanced, chemical pollutants still manage to slip into systems that people use every day. Substances such as trihalomethanes (THMs), bromides, and traces of heavy metals can make their way into tap water through everything from old pipes and industrial discharge to agricultural runoff after heavy rain.<a href="#ref1">¹</a> Repeated exposure, especially at low but consistent levels over time, has been linked with chronic diseases, including several types of cancer such as leukemia and colorectal cancer. The U.S. Environmental Protection Agency (EPA, 2024) has made it clear that ensuring safe drinking water is not just a matter of meeting the numbers on a chart but also about reducing toxic exposure that can silently build up over years.<a href="#ref2">²</a>  

In the context of San Diego, although the water supply of this city generally meets federal standards, that does not mean that it's perfectly safe in the long run. Those standards reflect only the latest findings about long-term exposure risks, some researchers point out. For example, chemicals such as THMs and bromides can turn problematic during long-term exposure. And here's where things could become problematic: depending on the source, treatment method, or even plumbing conditions in various neighborhoods, the concentration of these contaminants can vary considerably across town. A question that naturally follows: over time, have the fluctuations of those chemicals tracked with fluctuations in local cancer statistics? That may be a difficult connection to prove, yet investigating it offers insight at how environmental conditions and health outcomes interconnect in everyday life.  

There is some preliminary work that is worth noting. Ren, Martinez, and Cohen (2024) conducted a national-level analysis and found that areas with more Safe Drinking Water Act violations tended to report poorer physical and mental health—along with higher healthcare costs.<a href="#ref3">³</a> Liu, Zhang, and Wang (2024) looked at water pollution in China and found people with access to cleaner water supplies were far more likely to report their health as “very good.”<a href="#ref4">⁴</a> These are good studies that reinforce what seems intuitive but needs to be reinforced with data: water quality affects health. Yet, studies that link the specific chemistry of a city's water system—things like fluoride levels, THMs, and bromide—to concrete health outcomes, like cancer rates, still need to be done. By sifting through San Diego's water data over time and comparing it to public health statistics, this research hopes to find patterns that may go otherwise unseen. This research is important because it helps uncover how changes in San Diego’s water quality may be quietly affecting community health.

### Water Treatment Plant Effluent Chemical Parameters

This dataset contains measurements of chemical parameters from San Diego's water treatment plant effluents. The effluent data represents water quality after treatment but before distribution, providing insight into the effectiveness of treatment processes and the baseline quality of treated water entering the distribution system.

**Important Metrics and Units:**

- **Fluoride**: Measured in mg/L (milligrams per liter). The optimal range for drinking water fluoride is 0.7 mg/L according to CDC recommendations. Levels above 4 mg/L can cause dental fluorosis, while levels above 2 mg/L may pose health risks with long-term exposure.

- **Total THMs (Trihalomethanes)**: Measured in μg/L (micrograms per liter). THMs are disinfection byproducts formed when chlorine reacts with organic matter. The EPA maximum allowable level is 80 μg/L. THMs are associated with increased cancer risk and reproductive issues.

- **Bromide**: Measured in mg/L (milligrams per liter). While bromide itself is relatively non-toxic, it's a precursor to bromate formation during ozonation and can contribute to THM formation (which is associated with cancer risk and reproductive issues). Levels above 0.1 mg/L may be of concern in water treatment.

- **Chlorate and Chlorite**: Measured in mg/L (milligrams per liter). These are disinfection byproducts. EPA health advisory levels are 0.21 mg/L for chlorate and 0.2 mg/L for chlorite for infants and children.

- **pH**: Dimensionless logarithmic scale from 0-14. 0 is the most acid, and 14 is the most alkaline, with a 1 point difference indicating 10 times more acidity/alkalinity. Optimal drinking water pH should be between 6.5-8.5 according to EPA secondary standards. pH affects corrosivity and disinfection effectiveness.

**Major Concerns with the Dataset:**

This dataset has several limitations that could affect our analysis. First, the sampling locations are limited and not evenly distributed across San Diego, potentially creating geographic bias in our understanding of water quality. The dataset focuses primarily on treatment plant effluents, which may not reflect the actual quality of water reaching consumers due to distribution system effects. Additionally, some important chemical parameters like lead, copper, and certain organic contaminants are not included, limiting our ability to assess comprehensive water quality. The dataset also has very few measurements on key contaminants, such as bromate and nitrate, which may make those chemicals impossible to analyse with any significance.

The value_qualifier column has two unique values: 'ND' (Not Detected) and NaN (missing). 
As per the dictionary, 'ND' means not detected, and that is reflected in the analyte_value column, where all 'ND' entries have an analyte_value of nan.

However, the oposite is not true: not all missing analyte_value entries have a value_qualifier of 'ND'. Some are just missing without any qualifier.

Therefore, we decided to drop all of the null entries in the analyte_value column that do not have a value_qualifier of 'ND'. The ones with 'ND' will be filled with 0, since they were specifically said to be not detected.

We can also then remove the value_qualifier column since we won't need it anymore.

It is important to note that none of the null entries were in pH, so we are not inadvertently introducing bias by artificailly setting some samples to be really acidic. Values measured in MG/L or UG/L should be 0 if nothing was detected.

```python
df_effluent_clean = df_effluent.copy()
df_effluent_clean = df_effluent_clean.dropna(subset=['analyte_value', 'value_qualifier'], how='all')
df_effluent_clean = df_effluent_clean.drop('value_qualifier', axis=1)
df_effluent_clean['analyte_value'] = df_effluent_clean['analyte_value'].fillna(0)


print('Shape after clean: ', df_effluent_clean.shape)
print('Number of observations after clean:', df_effluent_clean.shape[0])
print('Number of variables after clean:', df_effluent_clean.shape[1])
df_effluent_clean.head()

```

```python
print("Null values after clean:")
print(df_effluent_clean.isnull().sum())
```

We need to graph each analyte to see their distributions and find any outliers.

```python
analytes = df_effluent_clean['analyte'].unique()
print(f"Available analytes: {analytes}")

fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(18, 20))
axes = axes.flatten()

for i, analyte in enumerate(analytes):
    analyte_data = df_effluent_clean[df_effluent_clean['analyte'] == analyte]

    axes[i].scatter(analyte_data['date_sample'], analyte_data['analyte_value'], alpha=0.6)
    axes[i].set_title(f'{analyte} over Time')
    axes[i].set_xlabel('Date Sampled')
    axes[i].set_ylabel('Analyte Value')

plt.tight_layout()
plt.show()
```

In looking at the plot, it is clear that only 5 analytes have significant data points: Fluoride, pH, Chlorate, Chlorite, and Total THMs. The rest have very few data points and are not worth analyzing. We will remove them from the dataset.

```python
analytes_to_keep = ['Fluoride', 'pH', 'Chlorate', 'Chlorite', 'Total THMs']
df_effluent_clean = df_effluent_clean[df_effluent_clean['analyte'].isin(analytes_to_keep)].copy()

print('Shape after clean: ', df_effluent_clean.shape)
print('Number of observations after clean:', df_effluent_clean.shape[0])
print('Number of variables after clean:', df_effluent_clean.shape[1])
print(df_effluent_clean['analyte'].value_counts())
df_effluent_clean.head()
```

There are a few obvious outliers that we have to remove.

```python

df_effluent_no_outliers = df_effluent_clean.copy()

print("Shape before removing outliers:", df_effluent_no_outliers.shape)


# Remove pH values > 14
ph_outliers = (df_effluent_no_outliers['analyte'] == 'pH') & (df_effluent_no_outliers['analyte_value'] > 14)
df_effluent_no_outliers = df_effluent_no_outliers[~ph_outliers]

# Remove Chlorate values >= 500
chlorate_outliers = (df_effluent_no_outliers['analyte'] == 'Chlorate') & (df_effluent_no_outliers['analyte_value'] >= 500)
df_effluent_no_outliers = df_effluent_no_outliers[~chlorate_outliers]

# Remove Fluoride values == 0
fluoride_outliers = (df_effluent_no_outliers['analyte'] == 'Fluoride') & (df_effluent_no_outliers['analyte_value'] == 0)
df_effluent_no_outliers = df_effluent_no_outliers[~fluoride_outliers]

print("Shape after removing outliers:", df_effluent_no_outliers.shape)

analytes = df_effluent_no_outliers['analyte'].unique()
print(f"Available analytes: {analytes}")

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 8))
axes = axes.flatten()

for i, analyte in enumerate(analytes):
    analyte_data = df_effluent_no_outliers[df_effluent_no_outliers['analyte'] == analyte]

    axes[i].scatter(analyte_data['date_sample'], analyte_data['analyte_value'], alpha=0.6)
    axes[i].set_title(f'{analyte} over Time')
    axes[i].set_xlabel('Date Sampled')
    axes[i].set_ylabel('Analyte Value')

plt.tight_layout()
plt.show()
```

### Water Distribution System Select Chemical Parameters

This dataset contains measurements of select chemical parameters from San Diego's water distribution system. The distribution system data represents water quality as it is delivered to consumers from the treatment plants. This provids insight into the actual water people are drinking past the treatment plants effluent stage as it travels through the pipes and infrastuctures.

**Important Metrics and Units:**

- **Fluoride**: Measured in mg/L (milligrams per liter). The optimal range for drinking water fluoride is 0.7 mg/L according to CDC recommendations. Levels above 4 mg/L can cause dental fluorosis, while levels above 2 mg/L may pose health risks with long-term exposure.
- **TON (Threshold Odor Number)**: Measured in ODOR units. TON is a measure of the odor intensity of water, with higher values indicating stronger odors. Values above 3 are generally considered objectionable for drinking water.
- **Color**: Measured in (COLOR units). Color in water is often due to dissolved organic matter, metals, or other substances.
- **Turbidity**: Measured in NTU (Nephelometric Turbidity Units). Turbidity indicates the cloudiness or haziness of water caused by suspended particles. The EPA recommends turbidity levels below 1 NTU for drinking water, as higher levels can harbor pathogens and reduce disinfection effectiveness.

**Major Concerns with the Dataset:**

- There are very few analytes we can look at, with only fluoride, color, TON, and turbidity.
- Many of the analytes such as turbidity, color, and TON have limited health implications compared to others not included in the dataset, such as lead, copper, or disinfection byproducts like trihalomethanes (THMs).
- The units given for the color analyte is unclear and ambiguous, making it difficult to interpret the measurements accurately. COLOR units are not a standard unit of measurement for water color, which is typically measured in Platinum-Cobalt units (Pt-Co) or Hazen units.
- Similarly to dataset #1, there are few locations of data collection distributed unevenly across San Diego, creating potential for geographic bias in our understanding of water quality.

### San Diego County Non-Communicable Cancer Outcomes (2016 - 2022)  

This dataset contains aggregated public health records for non-communicable cancers in the San Diego county between 2016 and 2022.


**Important Metrics and Units:**

- **Year**: Calendar year of the record (2016 - 2022)

- **Condition**: Cancer type, such as bladder cancer, lung cancer, or leukemia

- **Outcome**: The type of event that is being recorded and counted, usually Death or Hospitalization 

- **Geography / GeoType / GeoName / District / Region / SES**: Geographic and socioeconomic information describing where patients live (such as city name, supervisorial district, regional group, SES level)

- **AgeGroup**: A band/division of ages, ranging from 0-14, 15-24, etc.

- **Total**: The raw count of events for that condition, such as number of deaths

- **Rate**: The corresponding rate per 100,000 population or age-adjusted rate. This lets us compare locations of different sizes

**Major Concerns with the Dataset:**

This dataset has several limitations that could affect our analysis. First, the cancer outcome records are collected by geographic units such as cities, regions, and districts. Because these regions are uneven in population size, it risks creating geographical bias when we compare it across all locations. A worry we have is that small cities or regions with less residents may show large fluctuations in rates due to small event counts, rendering our rows statistically unstable. Also, the dataset tends to focus heavily on reported deaths and hospitalizations, but we recognize that this fails to account for areas with limited access to healthcare, fewer screening services, or lower engagement with statisticians. This could potentially result in underreporting bias. The dataset also lacks in substantial data, mosstly during the 2016-2019 years. Missing total counts or missing rate values is an indication of incomplete reporting or incocnsistent data collection procedures, all gaps that lower our ability to accurately assess long-term trends.

```python
import pandas as pd
import glob
import os

# These are the final clean names our dataset will use.
# Any cancer values not matching these will be removed or normalized
TARGET_CONDITIONS = [
    "Bladder Cancer",
    "Brain Cancer",
    "Colorectal Cancer",
    "Kidney Cancer",
    "Leukemia",
    "Liver Cancer",
    "Lung Cancer",
    "Non-Hodgkins Lymphoma",
    "Non-melanoma Skin Cancer",
    "Overall Cancer",
    "Pancreatic Cancer",
    "Prostate Cancer",
    "Thyroid Cancer"
]

# Standardizes messy cancer condition text into consistent labels.
# Also removes irrelevant parts like "non-communicable" or "death".
def clean_condition(cond):
    # If the input is not a string, we cannot clean it
    if not isinstance(cond, str):
        return None

    # Normalize the text for easier matching
    text = cond.lower()

    # These are the words to remove from the raw condition labels
    remove_words = [
        "non-communicable",
        "non communicable",
        "noncommunicable",
        "death",
        "in-patient treatment",
        "in patient treatment",
        "hospitalization",
        "incidence",
        "total"
    ]
    for w in remove_words:
        text = text.replace(w, "")

    # This line removes extra white spaces after cleaning
    text = text.strip()

    # This is a mapping of messy names → standardized cancer names
    mapping = {
        "bladder cancer": "Bladder Cancer",
        "brain cancer": "Brain Cancer",
        "colon and rectum cancer": "Colorectal Cancer",
        "colorectal cancer": "Colorectal Cancer",
        "kidney cancer": "Kidney Cancer",
        "leukemia": "Leukemia",
        "liver cancer": "Liver Cancer",
        "lung cancer": "Lung Cancer",
        "non-hodgkin lymphoma": "Non-Hodgkins Lymphoma",
        "non hodgkin lymphoma": "Non-Hodgkins Lymphoma",
        "non-hodgkin's lymphoma": "Non-Hodgkins Lymphoma",
        "non melanoma skin cancer": "Non-melanoma Skin Cancer",
        "skin cancer (non-melanoma)": "Non-melanoma Skin Cancer",
        "overall cancer": "Overall Cancer",
        "pancreatic cancer": "Pancreatic Cancer",
        "prostate cancer": "Prostate Cancer",
        "thyroid cancer": "Thyroid Cancer"
    }

    # Match cleaned text to one of the standardized names
    for key, val in mapping.items():
        if key in text:
            return val

    # If there are no matches found, then we return None
    return None


# This section is in particular to load/clean data from the 2016-2019 range
old_dataframes = []

#This loops through EVERY csv file (all 52) in the 2016-2019 folder
for file in glob.glob("./data/00-raw/2016_2019/*.csv"):
    print("Loaded:", file)

    # Read the raw CSV
    df = pd.read_csv(file)

    # Standardize the "CONDITION" column using our cleaning function from above
    df["Condition"] = df["CONDITION"].apply(clean_condition)

    # This keeps only the rows with one of the valid conditions
    df = df[df["Condition"].isin(TARGET_CONDITIONS)]

    # If nothing matched in this file, then we skip it
    if df.empty:
        continue

    # Some of the early datasets store cancer rate under "TotalRate"
    df["Rate"] = df.get("TotalRate")

    # Here, we create a simplified uniform scheme of values for 2016-2019
    # It is worth nothing that early datasets did not have age groups
    cleaned = pd.DataFrame({
        "Year": df["Year"],
        "Condition": df["Condition"],
        "Outcome": df["OUTCOME"],  
        "Geography": df["Geography"],
        "Region": df["Region"],
        "GeoType": df["GeoType"],
        "GeoName": df["GeoName"],
        "SES": df["SES"],
        "District": df["District"],
        "AgeGroup": None,    
        "Total": df["Total"],
        "Rate": df["Rate"]
    })

    old_dataframes.append(cleaned)

# Here, we need to combine all the data from the earlier years (before they changed their csv format)
df_old = pd.concat(old_dataframes, ignore_index=True)


# This section is in particular to load/clean data from the 2020-2022 range
new_dataframes = []

for file in glob.glob("./data/00-raw/2020_2022/*.csv"):
    print("Loaded:", file)
    df = pd.read_csv(file)

    # If we have any files that are missing the "CONDITION" column, we immediately skip over it
    if "CONDITION" not in df.columns:
        print("Skipped (no CONDITION column):", file)
        continue

    # Here we are standardizing the cancer names using the dictionary made above
    df["Condition"] = df["CONDITION"].apply(clean_condition)

    # Here we are keeping only the relevant conditions we mentioned above for cleaner data
    df = df[df["Condition"].isin(TARGET_CONDITIONS)]

    # When the dataframe is empty, it means that nothing was found there at all, and thus we continue
    if df.empty:
        print("No target conditions found in:", file)
        continue

    # This block of code is intended to identify the "Index Group" column, and makes sure we only have
    # rows where Index Group = "Total"
    index_col = None
    for possible in ["Index Group", "Index_Group", "IndexGroup", "Index"]:
        if possible in df.columns:
            index_col = possible
            break
    if index_col is not None:
        df = df[df[index_col] == "Total"]
    else:
        print("No Index Group column found in:", file)

    # This block is used to identify which column contains the number of case counts for data from 2020-2022
    count_col = None
    for possible in ["Count", "Counts", "Total", "Total_Count"]:
        if possible in df.columns:
            count_col = possible
            break

    if count_col is None:
        print("No count column found in:", file)
        continue

    # This is used to identify which column contains the rate for cancer(s)
    rate_col = None
    for possible in ["Rates", "Rate", "Age Adjusted Rate", "TotalRate"]:
        if possible in df.columns:
            rate_col = possible
            break

    # This creates a consistent and clean structure for the data from 2020-2022
    # It is noteworthy that AgeGroup is not available in the dataset
    cleaned = pd.DataFrame({
        "Year": df.get("Year"),
        "Condition": df.get("Condition"),
        "Outcome": df.get("OUTCOME"),
        "Geography": df.get("Geography"),
        "Region": df.get("Region"),
        "GeoType": df.get("GeoType"),
        "GeoName": df.get("GeoName"),
        "SES": df.get("SES"),
        "District": df.get("District"),
        "AgeGroup": None,
        "Total": df[count_col],
        "Rate": df[rate_col] if rate_col else None
    })

    new_dataframes.append(cleaned)

# Now we combine all the data from the later years (2020 - 2022)
df_new = pd.concat(new_dataframes, ignore_index=True)


# This is a final, clean merge of the cleaned dataframes from 2016-2019 and 2020-2022
final = pd.concat([df_old, df_new], ignore_index=True)

# This is standard unified column order
COLUMN_ORDER = [
    "Year",
    "Condition",
    "Outcome",
    "Geography",
    "Region",
    "GeoType",
    "GeoName",
    "SES",
    "District",
    "AgeGroup",
    "Total",
    "Rate"
]

# Here, we make sure that missing columns still appear in the final dataset
for col in COLUMN_ORDER:
    if col not in final.columns:
        final[col] = None

# And here, we are reordering the dataset into the proper standard unified column order
final = final[COLUMN_ORDER]

# Final, final cleanup of the total unified dataframes
print(final.info())
print("\nMissing values:\n", final.isna().sum())
print("\nDuplicate rows:", final.duplicated().sum())

final = final.replace(r'^\s*$', pd.NA, regex=True)
final = final.dropna(subset=["Total", "Rate"], how="all")
```

## Ethics

The data we are using does not unfortunately represent every neighborhood in San Diego; it’s biased towards more developed neighborhoods, meaning that some other communities within San Diego may be underrepresented. The main issue we find with the data we proposed is that, despite being generally good at showcasing the state of San Diego’s water quality as a whole city, it is not good at accurately representing the disparities in water quality across all San Diego residents.An issue that comes about from biases when sampling is that the health dataset is only representative from those who chose or were even able to seek medical care, meaning that certain communities or individuals with limited access to healthcare are going to be underrepresented in the data. And, unfortunately, this portion of the population would also most likely be the ones who similarly have limited access to cleaner water and thus may face greater environmental/health risks overall, risking our data’s integrity. Our data may be good at representing the whole population, but it is at risk of overrepresenting the San Diego residents with better access to healthcare and clean water.
To address the issues regarding bias and problematic data collection, we plan to detect and minimize it all the most before the data analysis. After compiling the locations, we will compare it to demographic data (such as income level and healthcare access) to find gaps within the data samples. After identifying if it makes an impact, we’ll communicate our findings transparently in our report. We will acknowledge which populations are overrepresented or underrepresented, use data visualizations to make it clear what these limitations are and how it affects our findings. This ensures that, regardless of the sample biases, our findings are accurate and respectful towards all communities represented within the data. With these two datasets, it does risk unintentionally exposing sensitive information about smaller, more vulnerable populations. We also plan to clearly communicate that differences in our findings can be attributed to external factors (such as outdated water infrastructure, uneven access to healthcare, etc) rather than the characteristics of residents in those areas.
