## Research Question

Has the introduction and availability of general-purpose AI like Chat-GPT influenced high school students’ academic performance in the United States?  And if so, how? Which measure of academic (GPA, Standadized test scores) performance was most affected by the introduction of AI/LLMs?




## Background and Prior Work

The arrival of generative AI tools–led by the likes of ChatGPT–has certainly remade the spaces of education among American high-schoolers. These tools, with instant access to explainer content, tutoring support, and writing assistance, raise questions about their widespread availability and how that could influence academic output. Preliminary indications suggest students increasingly incorporating AI into study habits. For example, surveys reveal that teen use of ChatGPT for homework assistance more than doubled from 2023 to 2024, rising from 13% to 26%<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). On the other hand, instructional stakeholders are responding to these developments: educators and institutions are launching initiatives to train teachers in AI usage and reevaluate assessment practices in the context of AI’s rising role. Also, in their recent COGS 108 project on college rankings and news story sentiment, they explored how technology interventions influence educational indicators, where they examined relationships among school access to resources, media attention, and performance data<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). Through their exploration, they found–though in another context–highlighted how emerging information technologies influence measurable academic performance through indirect motivational and behavioral processes. Previous studies have started to explore in greater detail how education is impacted by AI.

For example, Montenegro-Rueda et a. (2023) created a systematic review of students on the use of ChatGPT in learning environments and brought general positive effects on learning processes; although noting that studies on long-term performance results are still limited<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). Equally, Yan et al. (2023) surveyed 118 education articles on AI and noted that while AI facilitates personalized learning and feedback, it also poses questions regarding education preparedness, equity of access, and measures of outcomes <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). These findings underscore that although AI tools are designed to support students and teachers in their learning process, strong empirical tests of performance impacts–with special attention to high-school level on standardized test measures–are still nascent. 

Given this context, our project attempts to fill a gap by examining high-school level test scores (e.g. SATs, AP scores) vs. accessible use of AI products (e.g. ChatGPT). We aim at a U.S. time frame around the country-wide rollout of generative AI in schools and hope to generate more focused, contextual quantitative findings than currently reported in academic research.

## Hypothesis


After the introduction of AI/LLMs, like Chat-GPT, there has been a negative impact on students' academic performance in the United States, seen in standardized tests like SAT and AP tests. 

Null hypothesis: The introduction of general-purpose AI tools/LLMs have no significant effect on high school students’ performance within the United States. There has been no measurable/quantitative change in standardized test scores before/after the introduction of AI.

Alternative hypothesis: After the introduction of AI/LLMs, students' academic performance has had a positive impact, seen in overall increase in students scores in standardized tests in the United States.

### What/who would the observations be? Over what time period? etc.

Observe per subject (All English vs All Math vs All Science), to see if different subjects are affected differently
Observe per different test section (SAT math vs SAT english), to see if different subjects are affected differently, in similarly flavored tests
Observe per different test (ACT vs SAT vs AP), to see if different test structures are affected differently
Compare for survey data of students who use AI, as well as other collaborating factors (age/location/amount of use ) to see if there are specific groups that are more impacted by AI

These would all be observed over time, within the last 10-15 years, to compare if they have significantly changed since 2023 (as ChatGPT, what we consider the “start” of AI, was launched in November 2022)
